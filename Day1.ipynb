```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('diabetes.csv')
y=np.array(df[['BloodPressure']])
x=np.array(df[['Age']])
plt.scatter(x,y)
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
x_mean,y_mean
```

## To calculate the value of coefficient and intercept

```python
num=0
den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)*(x[i]-x_mean)
b1=num/den
b0=y_mean-(b1*x_mean)
b0
```

```python
y_pred= b0+b1*x
y_pred
```

```python
plt.scatter(x,y,c="red")
plt.plot(x,y_pred)
plt.show()
```
